# Chirp Signal

```julia
using PyPlot
```

```julia
c = 343; # Speed of sound in air in m/s
fs = 340*44100; # This is the sample rate of the sonar.
dt = 1/fs; # This is the sample spacing
r_max = 4; # Maximum range in metres to which to simulate.
t_max = 4E-3; # Time delay to max range

t = collect(0:dt:t_max); # t=0:dt:t_max defines a “range”.
print(t_max)
# Create an array containing the range values of the samples
r = c*t/2;
f0 = 40000; # Centre frequency is 10 kHz
B = 2000; # Chirp bandwidth
T = 4E-3; # Chirp pulse length
K = B/T; # Chirp rate
rect(t) = (abs.(t) .<= 0.5)*1.0
v_tx = 1*cos.( 2*pi*(f0*(t.-0.002 ) + 0.5*K*(t.-0.002 ).^2) ) .* rect.((t.-0.002)/T); 

print()

figure()
plot(v_tx[500:600],".")
print(size(v_tx))
```

```julia
print(round.(Int,v_tx .* 0.95*((2^(12))/2) .+ 0.95*((2^(12))/2)))
```

```julia
v_tx=(round.(Int,v_tx .* 0.99*((2^(12))/2) .+ 0.99*((2^(12))/2)))
figure()
plot(v_tx[500:600],".")
print(size(v_tx))
```
